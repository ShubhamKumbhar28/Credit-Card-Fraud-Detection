**As we can see the data is Imbalanced 
When dealing with imbalanced datasets, where the classes are not represented equally, the performance of machine learning models can be significantly affected. The model might become biased towards the majority class, leading to poor classification performance on the minority class.**

**Here are some strategies to handle imbalanced datasets**

**a. Oversampling Minority Class**
- Increase the number of instances in the minority class by replicating them to balance the dataset.

**b. Undersampling Majority Class**
- Reduce the number of instances in the majority class to balance the dataset.

**c. SMOTE (Synthetic Minority Over-sampling Technique)**
- Generate synthetic examples of the minority class using the SMOTE algorithm. This is a popular approach because it adds more    information to the model, helping it to generalize better.

```python
X=df.drop(['Class'],axis=1)
y=df['Class']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X_res,y_res,test_size=0.2,random_state=42)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
y_pred1 = log.predict(X_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
recall_score(y_test,y_pred1)
```

```python
f1_score(y_test,y_pred1)
```

```python
precision_score(y_test,y_pred1)
```
